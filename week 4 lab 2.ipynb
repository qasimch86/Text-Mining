## Sentiment Analysis using Naive Bayes

```python
#creating bag of words
from sklearn.feature_extraction.text import CountVectorizer 
vectorizer =CountVectorizer(stop_words='english')
```

```python
type(list_data)
```

```python
#converting the target or class to pandas
import pandas as pd
class_text=pd.Series(list_class)
class_text
```

```python
#transforming my text into pandas
import pandas as pd
df=pd.DataFrame(vectorizer.fit_transform(list_data).toarray(),columns=vectorizer.get_feature_names())
```

```python
df
```

```python
#for spliting your data into training and test set 
from sklearn.model_selection import train_test_split
#for analyzing the result of your model
from sklearn import metrics
#the two naive bayes algorithms
from sklearn.naive_bayes import GaussianNB 
from sklearn.naive_bayes import MultinomialNB
```

```python
# X_train, X_test, y_train, y_test = train_test_split(  X, y, test_size='', random_state=)
x_train,x_test,y_train,y_test=train_test_split(df,class_text,test_size=0.25,random_state=5)
```

```python
#defining your algorithm i.e. creating the object
algorithm_a=MultinomialNB()
```

```python
#model.fit(data,target)
#training your data
algorithm_a.fit(x_train,y_train)
```

```python
#testing your model with the test data
output=algorithm_a.predict(x_test)
```

```python
#analyzing your result using accuracy
metrics.accuracy_score(y_test,output)
```

```python
#analyzing your result using confusion matrix
metrics.confusion_matrix(y_test,output)
```
